# ***Data Cleaning***

```python
df.isna().sum()
```

```python
df.drop(columns = ['Status', 'Notes', 'Agency'], inplace = True, axis = 1)
df.head()
```

```python
#check duplicates values in data
df.duplicated().sum()
```

```python
df.rename(columns={'TotalPay': 'Salaries'}, inplace=True)
df.columns
```

```python
df.fillna(df.mean, inplace=True)
```

```python
df.isnull().sum()
```

# Basic Data ***Visualization***

```python
# distribution of salaries
df['Salaries'].hist(bins=20, edgecolor='black', figsize=(10, 6))
plt.xlabel('Salaries')
plt.ylabel('num of hits')
plt.show()
```

```python
# proportion of employees in different departments.
department_counts = df['JobTitle'].value_counts()
department_counts
```

```python
# we want to extract department from the job title column
df['Department'] = df['JobTitle'].str.extract(r'\((.*?)\)')
```

```python
df['Department'].head()
```

```python
# calculate the number of employees in each department
department_counts = df['Department'].value_counts()
department_counts
```

```python
# Plot a pie chart
plt.figure(figsize=(8, 8))
plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%', startangle=140)
plt.legend(title='Departments', labels=department_counts.index, bbox_to_anchor=(1, 0.5), loc='center left')
plt.title('Proportion of Employees in Different Departments')
plt.show()
```

# ***Grouped Analysis***

### grouping by depatment

```python
# nums of Agency
df['Department'].value_counts()
```

```python
grouped_by_Department = df.groupby('Department')['Salaries'].mean().sort_values(ascending = False).reset_index()
grouped_by_Department
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Salaries', y='Department', data=grouped_by_Department, palette='viridis')
plt.title('Mean Salaries by Department')
plt.xlabel('Mean Salary')
plt.ylabel('Department')
plt.show()
```

### grouping by year

```python
# range of years
df['Year'].value_counts()
```

```python
grouped_by_Year = df.groupby('Year')['Salaries'].mean().sort_values(ascending = False).reset_index()
grouped_by_Year
```

```python
plt.figure(figsize=(10, 6))
plt.bar(grouped_by_Year['Year'], grouped_by_Year['Salaries'], color='skyblue')
plt.title('Mean Salaries by Year')
plt.xlabel('Year')
plt.ylabel('Mean Salary')
plt.show()
```

# ***Simple Correlation Analysis***

# Identify any correlation between salary and another numerical column, and plot a scatter plot to visualize the relationship.

```python
df2 = df.copy()
```

```python
df2.columns
```

```python
df2.drop(columns = ['Id', 'EmployeeName', 'JobTitle', 'Year', 'Department'], inplace = True, axis = 1)
```

```python
print(df2.dtypes)
```

```python
conv_cols = ['BasePay', 'OvertimePay', 'OtherPay','Benefits']
df2[conv_cols] = df2[conv_cols].apply(pd.to_numeric, errors='coerce')
```

```python
df2.corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df2.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
scattered = ['BasePay', 'TotalPayBenefits', 'Benefits']

for i in scattered:
    sns.scatterplot(x=i, y='Salaries', data=df2)
    plt.show()
```
